```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
start = 0
end = 1
w_step = 1
w_stop = 100
x_step = 0.01
#p_step = 0.5 * math.pi * np.arange(w_stop // w_step)
phase = np.random.rand(w_stop // w_step) * 2 * math.pi

xs = np.arange(start, end, x_step)
ws = 2*math.pi*np.arange(0, w_stop, w_step)

sines = np.sin(ws[:, None] * xs[None, :] + phase[:, None])

# TODO normalize the signal and round into 8 bins between [-1, 1]
# TODO how do we estimate the probabilities?
# - quantize and count
# - use a gaussian mixture model or some other density estimator (we need to normalize by the )
# - the density will go to zero rapidly outside the domain and since the domain is [-1, 1]
#   we normalize by the volume of the n-dimensional hypercube (namely 2**n)

sine_sums = np.sum(sines[:10,:], axis=0)
sine_sums /= np.abs(sine_sums).max()
bins = np.arange(-1, 1.25, 0.25)
sine_sums = np.digitize(sine_sums, bins)
plt.plot(xs, sine_sums)
```
